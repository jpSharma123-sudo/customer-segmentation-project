## Step 1 — Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import warnings
warnings.filterwarnings('ignore')

# Consistent plot style
sns.set_style('darkgrid')
plt.rcParams['figure.figsize'] = (10, 6)
COLORS = ['#38bdf8', '#34d399', '#f472b6', '#a78bfa', '#fb923c']
print('Libraries imported successfully')
```

## Step 2 — Load & Inspect Data

```python
# Load dataset
df = pd.read_csv('Mall_Customers.csv')

print('Shape:', df.shape)
print('\nColumn Names:', df.columns.tolist())
df.head(10)
```

```python
# Data types and basic info
print(df.info())
print(df.shape)
print(df.describe())
```

## Step 3 — Preprocessing & Missing Value Check

```python
# Check missing values
print('Missing Values:')
print(df.isnull().sum())

# Encode Gender for correlation analysis
df_encoded = df.copy()
df_encoded['Gender'] = df_encoded['Gender'].map({'Male': 0, 'Female': 1})

# Drop CustomerID — not a feature
df_encoded.drop(columns=['CustomerID'], inplace=True)

print('\nStatistical Summary:')
df.describe().round(2)
```

## Step 4 — Exploratory Data Analysis (EDA)

### A. Feature Distribution

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
cols = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']

for ax, col, color in zip(axes, cols, COLORS):
    ax.hist(df[col], bins=20, color=color, alpha=0.8, edgecolor='white')
    ax.set_title(f'Distribution of {col}', fontweight='bold')
    ax.set_xlabel(col)
    ax.set_ylabel('Frequency')

plt.suptitle('Feature Distributions', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

### B. Income vs Spending Score by Gender Age vs Spending Score 

```python
# Income vs Spending Score coloured by Gender
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# By gender
for gender, color in zip(['Male','Female'], COLORS):
    subset = df[df['Gender'] == gender]
    axes[0].scatter(subset['Annual Income (k$)'], subset['Spending Score (1-100)'],
                    c=color, label=gender, alpha=0.8, s=70)
axes[0].set_title('Income vs Spending Score (by Gender)', fontweight='bold')
axes[0].set_xlabel('Annual Income (k$)')
axes[0].set_ylabel('Spending Score')
axes[0].legend()

# Age vs Spending Score
axes[1].scatter(df['Age'], df['Spending Score (1-100)'], c=COLORS[3], alpha=0.7, s=70)
axes[1].set_title('Age vs Spending Score', fontweight='bold')
axes[1].set_xlabel('Age')
axes[1].set_ylabel('Spending Score')

plt.tight_layout()
plt.show()
```

### C. Outlier Detection Using Boxplots;  Gender Distribution;  Spending Score By Gender

```python
# Boxplots — detect outliers per feature
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
cols = ["Age", "Annual Income (k$)", "Spending Score (1-100)"]
for ax, col, color in zip(axes, cols, COLORS):
    ax.boxplot(df[col], patch_artist=True,
               boxprops=dict(facecolor=color, alpha=0.7),
               medianprops=dict(color="red", linewidth=2))
    ax.set_title(f"Boxplot: {col}", fontweight="bold")
    ax.set_ylabel(col)
plt.suptitle("Outlier Detection via Boxplots", fontsize=13, fontweight="bold")
plt.tight_layout()
plt.show()

# Gender distribution — pie chart
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

gender_counts = df['Gender'].value_counts()

axes[0].pie( gender_counts, labels=gender_counts.index, autopct='%1.1f%%',
             startangle=90,shadow=True,
             explode=[0.03, 0.03],  colors=['#1f77b4', '#d62728'])
axes[0].set_title(
    'Gender Distribution',
    fontsize=14,
    fontweight='bold'
)

# Spending Score by Gender
for gender, color in zip(['Male', 'Female'], ['#1f77b4', '#d62728']):
    data = df[df['Gender'] == gender]['Spending Score (1-100)']
    axes[1].hist( data,bins=15, alpha=0.75, label=gender,color=color,edgecolor='black' )
axes[1].set_title(
    'Spending Score by Gender',
    fontsize=14,
    fontweight='bold'
)
axes[1].set_xlabel('Spending Score')
axes[1].set_ylabel('Frequency')

axes[1].legend()
axes[1].grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

### D. Correlation Heatmap

```python
plt.figure(figsize=(7, 7))
corr = df_encoded.corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm',
            square=True, linewidths=0.5, vmin=-1, vmax=1)
plt.title('Feature Correlation Heatmap', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

## Step 5 — Feature Scaling

```python
# Select features for clustering
X = df[['Annual Income (k$)', 'Spending Score (1-100)']].values

# StandardScaler: mean=0, std=1
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print('Before Scaling — Mean:', X.mean(axis=0).round(2), '| Std:', X.std(axis=0).round(2))
print('After  Scaling — Mean:', X_scaled.mean(axis=0).round(4), '| Std:', X_scaled.std(axis=0).round(4))
```

## Step 6 — Find Optimal K: Elbow Method & Silhouette Score

```python
inertias, sil_scores = [], []
K_range = range(2, 11)

for k in K_range:
    km = KMeans(n_clusters=k, init='k-means++', random_state=42, n_init=10)
    km.fit(X_scaled)
    inertias.append(km.inertia_)
    sil_scores.append(silhouette_score(X_scaled, km.labels_))

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Elbow Method
axes[0].plot(list(K_range), inertias, 'bo-', linewidth=2, markersize=8)
axes[0].axvline(x=5, color='red', linestyle='--', label='K=5 (optimal)')
axes[0].set_title('Elbow Method — Inertia vs K', fontweight='bold')
axes[0].set_xlabel('Number of Clusters'); axes[0].set_ylabel('WCSS (Inertia)')
axes[0].legend()

# Silhouette
axes[1].plot(list(K_range), sil_scores, 'gs-', linewidth=2, markersize=8)
axes[1].axvline(x=5, color='red', linestyle='--', label='K=5 (optimal)')
best_k = list(K_range)[sil_scores.index(max(sil_scores))]
axes[1].set_title('Silhouette Score vs K', fontweight='bold')
axes[1].set_xlabel('Number of Clusters'); axes[1].set_ylabel('Silhouette Score')
axes[1].legend()

plt.tight_layout()
plt.show()
print(f' Optimal K = {best_k} | Silhouette Score = {max(sil_scores):.4f}')
```

## Step 7 — Apply K-Means (K=5)

```python
# Train final model
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(X_scaled)

# Inverse transform cluster centers to original scale
centers = scaler.inverse_transform(kmeans.cluster_centers_)

print(f'Final Silhouette Score: {silhouette_score(X_scaled, df["Cluster"]):.4f}')
print(f'Cluster Sizes:\n{df["Cluster"].value_counts().sort_index()}')
```

## Step 8 — Visualizing clusters

```python
seg_labels = {
    0: 'Average Customers',
    1: 'High Income High Spend (VIP)',
    2: 'Low Income High Spend (Impulsive)',
    3: 'High Income Low Spend (Cautious)',
    4: 'Low Income Low Spend (Budget)'
}

plt.figure(figsize=(11, 8))
for c, color in enumerate(COLORS):
    mask = df['Cluster'] == c
    plt.scatter(df.loc[mask, 'Annual Income (k$)'],
                df.loc[mask, 'Spending Score (1-100)'],
                c=color, s=80, alpha=0.85, label=f'Cluster {c}: {seg_labels[c]}')

plt.scatter(centers[:, 0], centers[:, 1],
            c='black', s=250, marker='*', label='Centroids', zorder=5)

plt.title('K-Means Customer Segments (K=5)', fontsize=14, fontweight='bold')
plt.xlabel('Annual Income (k$)', fontsize=12)
plt.ylabel('Spending Score (1-100)', fontsize=12)
plt.legend(loc='upper left', fontsize=9)
plt.tight_layout()
plt.show()
```

## Step 9 — Cluster Profiles & Segment Insights

### A. Aggregate cluster statistics

```python
profile = df.groupby('Cluster').agg(
    Count       = ('Cluster', 'count'),
    Avg_Age     = ('Age', 'mean'),
    Avg_Income  = ('Annual Income (k$)', 'mean'),
    Avg_Spend   = ('Spending Score (1-100)', 'mean')
).round(1)

profile['Segment'] = profile.index.map(seg_labels)
profile
```

### B. Bar chart comparison across clusters

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
metrics = [('Avg_Age', 'Average Age'), ('Avg_Income', 'Avg Annual Income (k$)'),
           ('Avg_Spend', 'Avg Spending Score')]

for ax, (col, title) in zip(axes, metrics):
    bars = ax.bar([f'C{i}' for i in range(5)], profile[col], color=COLORS, alpha=0.85, edgecolor='white')
    for bar, val in zip(bars, profile[col]):
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.3,
                f'{val:.1f}', ha='center', fontsize=9, fontweight='bold')
    ax.set_title(title, fontweight='bold')
    ax.set_xlabel('Cluster')

plt.suptitle('Cluster Profile Comparison', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

### C. Donut chart — Customer Segment Distribution

```python
seg_labels_short = {
    0: "Average Customers",
    1: "High Income High Spend (VIP)",
    2: "Low Income High Spend (Impulsive)",
    3: "High Income Low Spend (Cautious)",
    4: "Low Income Low Spend (Budget)"
}

cluster_counts = df["Cluster"].value_counts().sort_index()
labels = [f"C{i}: {seg_labels_short[i]}\n(n={cluster_counts[i]})" for i in range(5)]

fig, ax = plt.subplots(figsize=(9, 9))
wedges, texts, autotexts = ax.pie(
    cluster_counts,
    labels=labels,
    autopct="%1.1f%%",
    colors=COLORS,
    startangle=140,
    pctdistance=0.78,
    wedgeprops=dict(width=0.55, edgecolor="white", linewidth=2)
)
for at in autotexts:
    at.set_fontweight("bold")
    at.set_fontsize(10)

ax.set_title("Customer Segment Distribution (Donut Chart)",
             fontsize=14, fontweight="bold", pad=20)
plt.tight_layout()
plt.show()
```

### A. Save trained model and scaler for future use

```python
import joblib
import os

# Create models folder
os.makedirs("models", exist_ok=True)

# Save files
joblib.dump(kmeans, "models/kmeans_k5.pkl")
joblib.dump(scaler, "models/standard_scaler.pkl")

print("Model saved to models/kmeans_k5.pkl")
print("Scaler saved to models/standard_scaler.pkl")

# Reload and verify
loaded_model = joblib.load("models/kmeans_k5.pkl")
loaded_scaler = joblib.load("models/standard_scaler.pkl")

test_pred = loaded_model.predict(
    loaded_scaler.transform([[60, 50]])
)

print(f"Test prediction for Income=60k, Score=50 → Cluster {test_pred[0]}")
```
